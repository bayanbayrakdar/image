```python
from PIL import Image
import matplotlib.pyplot as plt
image=Image.open("lenna.png")
image.format
image.size
image.width
image.height
image.show()
image.mode
# plt.imshow(image)
type(image)
```

**Image Operations**

```python
from PIL import ImageOps
image_flip=ImageOps.flip(image)
image_flip.show()

image_mirror=ImageOps.mirror(image)
image_mirror.show
image_grayscale=ImageOps.grayscale(image)
image_grayscale.show()
image_invert=ImageOps.invert(image)
image_invert.show
image_grayscale.quantize(2)

## get three chanels
red,green,blue=image.split()

red.show()
green.show()
blue.show()
plt.subplot(1,3,1)
plt.title("red")
plt.imshow(red)
plt.subplot(1,3,2)
plt.title("blue")
plt.imshow(blue)
plt.subplot(1,3,3)
plt.title("green")
plt.imshow(green)
plt.show()

## convert image to numpy array
import numpy as np
array_image=np.array(red)
print(array_image)

```
