**Geometric Operation**
1. scaling
2. Translation: shift the image

3.Rotation

we change the coordinates of the image x and y

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
image=Image.open("cameraman.jpg")
plt.subplot(1,2,1)
plt.imshow(image)
width,height=image.size
new_image=image.resize((width*2,height))
plt.subplot(1,2,2)
plt.imshow(new_image)
```

```python
import cv2
image=cv2.imread("cameraman.jpg")
new_image=cv2.resize(image,(image.shape[1]*2,image.shape[0]))
plt.imshow(new_image)
new_image=cv2.resize(image,None,fx=2,fy=1,interpolation=cv2.INTER_CUBIC)
plt.imshow(new_image)
```

```python
rows,cols=image.shape[:2]
M=np.float32([[1,0,100],[0,1,50]])
new_image=cv2.warpAffine(image,M,(cols,rows))
plt.imshow(new_image)
```

```python
theta=45
M=cv2.getRotationMatrix2D((cols/2,rows/2),theta,1)
new_image=cv2.warpAffine(image,M,(cols,rows))
plt.imshow(new_image)
```
