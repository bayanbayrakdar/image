**Manipulating Image one pixel at a time**
1. cropping
2. changing Image pixels

```python
from PIL import Image
import numpy as np
import cv2
image = Image.open("lenna.png")
image=np.array(image)
upper=250
lower=300
crop_image=image[upper:lower,:,:]
crop_image
crop_horizintal=crop_image[:,200:400,:]
crop_horizintal
```

```python
#change pixel image
from PIL import ImageDraw
image_draw=image.copy()
image_draw=Image.fromarray(image_draw)
image_fn=ImageDraw.Draw(image_draw)
image_fn=image_fn.rectangle((200,200,400,400),fill="red")
plt.imshow(image_draw)
```

```python
IMAGE_BAR=Image.open("barbara.png")
IMAGE_BAR1=np.array(IMAGE_BAR)
IMAGE_BAR.paste(image_draw,box=(150,150))
plt.imshow(IMAGE_BAR)
```
